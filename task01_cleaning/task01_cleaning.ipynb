# Task 01 – Data Cleaning

Mục tiêu: Đọc và làm sạch dữ liệu shipper, order và rating

```python
import pandas as pd
import numpy as np
```

```python
import os
print(os.getcwd())
print(os.listdir('../data_raw'))
```

```python
shipper = pd.read_csv('../data_raw/shipper_info.csv')
order   = pd.read_csv('../data_raw/order_info.csv')
rating  = pd.read_csv('../data_raw/customer_rating.csv')

shipper.head()
```

```python
shipper['shipper_id'] = shipper['shipper_id'].astype(str).str.strip()
shipper['full_name'] = shipper['full_name'].str.strip().str.title()

shipper['region'] = (
    shipper['region']
    .str.strip()
    .str.lower()
    .replace({
        'bac': 'Bắc',
        'trung': 'Trung',
        'nam': 'Nam'
    })
)
```

```python
order['distance_km'] = (
    order['distance_km']
    .astype(str)
    .str.replace('[^0-9.-]', '', regex=True)
)

order['distance_km'] = pd.to_numeric(
    order['distance_km'],
    errors='coerce'
)

order.loc[order['distance_km'] < 0, 'distance_km'] = np.nan
```

```python
rating['rating'] = pd.to_numeric(rating['rating'], errors='coerce')
rating.loc[~rating['rating'].between(1, 5), 'rating'] = np.nan
```

```python
shipper.to_csv('../data_clean/shipper_clean.csv', index=False)
order.to_csv('../data_clean/order_clean.csv', index=False)
rating.to_csv('../data_clean/rating_clean.csv', index=False)
```
