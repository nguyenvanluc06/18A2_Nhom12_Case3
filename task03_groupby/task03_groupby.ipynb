# Task 03 – GroupBy & Tổng hợp dữ liệu

Mục tiêu: Phân tích hiệu suất giao hàng bằng các phép groupby

```python
import pandas as pd

shipper = pd.read_csv('../data_clean/shipper_clean.csv')
order = pd.read_csv('../data_clean/order_clean.csv')
rating = pd.read_csv('../data_clean/rating_clean.csv')
```

```python
data = (
    order
    .merge(shipper, on='shipper_id', how='left')
    .merge(rating, on='order_id', how='left')
)

data.head()
```

```python
avg_time_shipper = data.groupby('shipper_id')['delivery_time_min'].mean()
avg_time_shipper
```

```python
avg_distance_region = data.groupby('region')['distance_km'].mean()
avg_distance_region
```

```python
fastest_shipper = avg_time_shipper.idxmin()
fastest_shipper
```

```python
shipper_performance = data.groupby('shipper_id').agg({
    'delivery_time_min': 'mean',
    'rating': 'mean'
}).sort_values(
    by=['delivery_time_min', 'rating'],
    ascending=[True, False]
)

shipper_performance
```
