# Task 05 – Pivot Table

Mục tiêu: Tổng hợp dữ liệu bằng Pivot Table để phân tích hiệu suất giao hàng


```python
df = order.merge(shipper, on='shipper_id', how='left')
df = df.merge(rating, on='order_id', how='left')
```

```python
task5 = (
    df.groupby(['shipper_id', 'full_name', 'region'])
      .agg(
          total_orders=('order_id', 'count'),
          avg_distance_km=('distance_km', 'mean'),
          avg_delivery_time=('delivery_time_min', 'mean'),
          avg_rating=('rating', 'mean')
      )
      .reset_index()
)
```
