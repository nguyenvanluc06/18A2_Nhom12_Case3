# Task 04 – Merge dữ liệu

Mục tiêu: Kết hợp các bảng dữ liệu để tạo bảng phân tích hoàn chỉnh


```python
import pandas as pd

shipper = pd.read_csv('../data_clean/shipper_clean.csv')
order = pd.read_csv('../data_clean/order_clean.csv')
rating = pd.read_csv('../data_clean/rating_clean.csv')
```

```python
shipper_order = pd.merge(
    order,
    shipper,
    on='shipper_id',
    how='left'
)

shipper_order.head()
```

```python
final_data = pd.merge(
    shipper_order,
    rating,
    on='order_id',
    how='left'
)

final_data.head()
```

```python
final_data.info()
```

```python
final_data.to_csv('../data_clean/final_merged_data.csv', index=False)
```
