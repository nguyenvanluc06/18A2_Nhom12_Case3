# Task 02 – Query & Thống kê dữ liệu

Mục tiêu: Thực hiện các truy vấn và thống kê mô tả theo yêu cầu


```python
import pandas as pd

shipper = pd.read_csv('../data_clean/shipper_clean.csv')
order   = pd.read_csv('../data_clean/order_clean.csv')
rating  = pd.read_csv('../data_clean/rating_clean.csv')
```

```python
order_count = order.groupby('shipper_id')['order_id'].count()
order_count
```

```python
order_no_rating = order[~order['order_id'].isin(rating['order_id'])]
order_no_rating.head()
```

```python
avg_distance = order['distance_km'].mean()
avg_distance
```

```python
order_shipper = order.merge(
    shipper[['shipper_id', 'region']],
    on='shipper_id',
    how='left'
)

avg_time_region = order_shipper.groupby('region')['delivery_time_min'].mean()
avg_time_region
```
